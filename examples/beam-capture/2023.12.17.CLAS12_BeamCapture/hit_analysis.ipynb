```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('port7001.evio.csv', nrows=100000)  # 

df.head()
```

```python
plt.figure(figsize=(16,2))
plt.hist(df['slot'], bins=34, color='green', range=[-1.5, 32.5], log=True)
plt.title('FADC250 slots')
plt.xlabel('slota') 
plt.grid()
plt.show()

plt.figure(figsize=(16,2))
plt.hist(df['chan'], bins=18, color='orange', range=[-1.5, 16.5], log=True)
plt.title('FADC250 channels')
plt.xlabel('channel') 
plt.grid()
plt.show()

plt.figure(figsize=(16,2))
plt.hist(df['q'], bins=2048, color='red', range=[-0.5, 2047.5], log=True)
plt.title('FADC250 q')
plt.xlabel('q') 
plt.grid()
plt.show()

plt.figure(figsize=(16,2))
plt.hist(df['t'], bins=4096, color='blue', range=[-0.5, 65535.5], log=True)
plt.title('FADC250 t')
plt.xlabel('t(ns)') 
plt.grid()
plt.show()
```
