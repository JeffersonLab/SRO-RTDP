## Read in beam current tree

```python
from datetime import datetime
from array import array

# Define the reference date
reference_date = datetime.strptime("2024-05-16 09:00:00", "%Y-%m-%d %H:%M:%S")

# Create a ROOT TTree
ibeam_tree = ROOT.TTree("T", "A tree with time and beam current")

# Define the branches
time = array('f', [0])  # Branch for time in seconds (float)
ibeam = array('f', [0])  # Branch for current (float)

ibeam_tree.Branch("time", time, "time/F")
ibeam_tree.Branch("ibeam", ibeam, "ibeam/F")

# Read the text file
with open('/nvme/proj/RTDP/2024.05.17.CLAS12/sroTest_HallB_05162024_BeamCurrent.txt', 'r') as file:
    # Skip the header line
    next(file)
    
    for line in file:
        # Parse the line
        parts = line.split()
        date_str = parts[0] + " " + parts[1]
        value = float(parts[2])
        
        # Compute the time in seconds relative to the reference date
        current_date = datetime.strptime(date_str, "%Y-%m-%d %H:%M:%S")
        time[0] = (current_date - reference_date).total_seconds()
        ibeam[0] = value
        
        # Fill the tree
        ibeam_tree.Fill()

# Save the tree to a ROOT file
# output_file = ROOT.TFile("output.root", "RECREATE")
# tree.Write()
# output_file.Close()
```

```python
ibeam_tree.SetMarkerSize(1)
ibeam_tree.SetMarkerStyle(20)
ibeam_tree.Draw("ibeam:time")
```
